### Preliminaries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#seaborn settings
sns.set_palette("deep", desat=.6)
sns.set_context(rc={"figure.figsize": (8,4)})
```

### Load Data Sets

```python
titanic = pd.read_csv('data/titanic.csv')
```

```python
server = pd.read_csv('data/serverdata.csv')
```

```python
mlb = pd.read_csv('data/mlbsalaries.csv')
```

```python
titanic.head()
```

```python
server.head()
```

```python
mlb.head()
```

### 1: Histograms

```python
plt.hist(titanic.Age.dropna(), bins=35)
```

```python
plt.hist(server.Time, 25, color=sns.desaturate('indianred', 1))
```

```python
plt.hist(mlb.Salary.dropna())
```

### 2. Box Plots (Shows min, max, first quartile, median and third quartile)

```python
sns.boxplot(titanic.Age)
```

```python
sns.boxplot(titanic.Age, titanic.Sex)
```

```python
sns.boxplot(server.Time)
```

```python
sns.boxplot(mlb.Salary)
```

### 3. Kernel Density Plot

```python
sns.kdeplot(titanic.Age, shade=True)
```

```python
sns.kdeplot(server.Time, shade=True)
```

```python
sns.kdeplot(mlb.Salary, shade=True)
```

```python
sns.distplot(titanic.Age.dropna())
```

```python
sns.distplot(server.Time.dropna())
```

```python
sns.distplot(mlb.Salary.dropna())
```

### 4. Violon Plot

```python
sns.violinplot(titanic.Age.dropna())
```

```python
sns.violinplot(titanic.Age.dropna(), titanic.Sex)
```

```python
sns.violinplot(mlb.Salary.dropna())
```

```python
sns.violinplot(server.Time.dropna())
```

### 5. Cumulative Distribution Function

```python
sns.kdeplot(titanic.Age.dropna(), cumulative=True, shade=True)
```

```python
sns.kdeplot(serve..dropna(), cumulative=True, shade=True)
```
